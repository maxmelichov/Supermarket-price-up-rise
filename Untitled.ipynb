```python
def price_toCsv(place,filename):
    cols = ["PriceUpdateDate"
                ,"ItemCode"
                 ,"ItemType"
                 ,"ItemNm"
                 ,"ManufacturerName"
                 ,"ManufactureCountry"
                 ,"ManufacturerItemDescription"
                 ,"UnitQty"
                 ,"Quantity"
                  ,"UnitOfMeasure"
                 ,"bIsWeighted"
                 ,"QtyInPackage"
                 ,"ItemPrice"
                ,"UnitOfMeasurePrice"
                 ,"AllowDiscount"
                 ,"ItemStatus"
                ,"Location"]
    rows = []
# Parsing the XML file
    xml = ElementTree.parse(filename+'.xml')
    print(xml)
    root = xml.getroot()
    print(root)
#print(root.findall('./Items/Item'))
   
    for i in root.findall('./Items/Item'):
        if 'גבינה' in i.find("ItemNm").text:
            PriceUpdateDate = i.find("PriceUpdateDate").text
            ItemCode = i.find("ItemCode").text
            ItemType = i.find("ItemType").text
            ItemNm = i.find("ItemNm").text
            ManufacturerName = i.find("ManufacturerName").text
            ManufactureCountry = i.find("ManufactureCountry").text
            ManufacturerItemDescription = i.find("ManufacturerItemDescription").text
            UnitQty = i.find("UnitQty").text
            Quantity = i.find("Quantity").text
            UnitOfMeasure = i.find("UnitOfMeasure").text
            bIsWeighted = i.find("bIsWeighted").text
            QtyInPackage = i.find("QtyInPackage").text
            ItemPrice = i.find("ItemPrice").text
            UnitOfMeasurePrice = i.find("UnitOfMeasurePrice").text
            AllowDiscount = i.find("AllowDiscount").text
            ItemStatus = i.find("ItemStatus").text
            Location=place

            rows.append({"PriceUpdateDate": PriceUpdateDate,
                        "ItemCode": ItemCode,
                        "ItemType": ItemType,
                        "ItemNm": ItemNm,
                        "ManufacturerName": ManufacturerName,
                        "ManufactureCountry": ManufactureCountry,
                        "ManufacturerItemDescription": ManufacturerItemDescription,
                        "UnitQty": UnitQty,
                        "Quantity": Quantity,
                        "UnitOfMeasure": UnitOfMeasure,
                        "bIsWeighted": bIsWeighted,
                        "QtyInPackage": QtyInPackage,
                        "ItemPrice": ItemPrice,
                        "UnitOfMeasurePrice": UnitOfMeasurePrice,
                        "AllowDiscount": AllowDiscount,
                        "ItemStatus": ItemStatus,
                        "Location": Location})
    df = pd.DataFrame(rows, columns=cols)
    # Writing dataframe to csv
    df.to_csv('price.csv')
    df1=pd.read_csv('price.csv')
    df2=pd.read_csv('all_prices.csv')
    frames = [df1, df2]
    result = pd.concat(frames,ignore_index=True)
    result.to_csv('all_prices.csv')
    
```
